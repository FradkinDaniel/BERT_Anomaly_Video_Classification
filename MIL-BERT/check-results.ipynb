```python
import os
import numpy as np 
import random 
import glob 
```

```python
log1_files = glob.glob('*log1.txt')
log1_files.sort()
print(log1_files[:5])
print(len(log1_files))
```

```python
log2_files = glob.glob('*log2.txt')
log2_files.sort()
print(log2_files[:5])
print(len(log2_files))
```

```python
for f1 in log1_files:
    
    
    f2_s = f1.split('_')  
    f2_s[-1] = 'log2.txt' 
    #print(f2_s)
    f2 = '_'.join(f2_s)
  
    print(f1,f2)

    
    lines = open(f1,'r').readlines() 
    lines = [x.strip() for x in lines]
    gt1 = [float(x.split(' ')[0]) for x in lines]
    score1 = [float(x.split(' ')[1]) for x in lines]
    
    lines = open(f2,'r').readlines() 
    lines = [x.strip() for x in lines]
    gt2 = [float(x.split(' ')[0]) for x in lines]
    score2 = [float(x.split(' ')[1]) for x in lines]
    #check if any gts are different 
    for i,(g1,g2) in enumerate(zip(gt1,gt2)): 
        if g1!=g2:
            print("GTs are different at ", f1,f2,i)
    #check if any scores are different     
    for i,(s1,s2) in enumerate(zip(score1,score2)): 
        if s1!=s2:
            print("GTs are different at ", f1,f2,i)
            print(i,s1,s2)
            break 
    
    
    
```
